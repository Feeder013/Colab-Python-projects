```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing = pd.read_csv(r'/content/real_estate.csv', encoding='ISO-8859-1')
housing
```

```python
housing_filtered = housing[housing['state'].isin(['Puerto Rico','Massachusetts','New York'])]
```

```python
sns.set_theme(style='whitegrid')
sns.set_style('darkgrid')
sns.set_palette('deep')
sns.barplot(housing_filtered, x='state', y='price', estimator='median')
```

```python
housing_filt = housing[housing['house_size']< 10000]
sns.regplot(housing_filt, x='house_size', y='price', line_kws={'color':'red'})
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
housing_cfilt = housing.drop(columns=['status','zip_code','city','state','prev_sold_date'])
housing_cfilt
housing_cfilt.corr()
```

```python
spearman_correlation = housing_cfilt.corr(method='spearman')
spearman_correlation
pearson_correlation = housing_cfilt.corr(method='pearson')
pearson_correlation
kendall_correlation = housing_cfilt.corr(method='kendall')
kendall_correlation
```

```python
fig, axs = plt.subplots(3,1, figsize=(10,7))
sns.heatmap(spearman_correlation, ax=axs[0], annot=True, cmap='coolwarm')
axs[0].set_title('Spearman Correlation')
sns.heatmap(pearson_correlation, ax=axs[1], annot=True,cmap='coolwarm')
axs[1].set_title('Pearson Correlation')
sns.heatmap(kendall_correlation, ax=axs[2], annot=True,cmap='coolwarm')
axs[2].set_title('Kendall Correlation')
plt.tight_layout()
plt.show()
```
