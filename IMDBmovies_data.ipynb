```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv(r'/usr/imdb_movies.csv', index_col='id')
pd.set_option('display.max_rows',11000)
movies
movies.head()
```

```python
movies.describe()
```

```python
movies.drop_duplicates(inplace=True)
movies.head()
```

```python
movies.dropna(subset=['genres'], inplace=True)
movies['Profit'] = movies['revenue'] - movies['budget']
movies.head()
```

```python
#movies.drop(['cast','homepage', 'director','tagline','keywords'],axis=1)
movies.head()
movies[['main_genres','sub_genres1','sub_genres2','sub_genres3','sub_genres4']] = movies['genres'].str.split('|',expand=True)
movies.head()
```

```python
#movies.drop(['genres','sub_genres1','sub_genres2','sub_genres3','sub_genres4'], axis=1, inplace=True)
movies.head()
#movies[['main_companies','sub_companies1','sub_companies2','sub_companies3','sub_companies4']] = movies['production_companies'].str.split('|',expand=True)
#movies.drop(['production_companies','sub_companies1','sub_companies2','sub_companies3','sub_companies4'], axis=1, inplace=True)
```

```python
movies.head()
genre_count = movies['main_genres'].value_counts()
genre_countdf = genre_count.reset_index()
genre_countdf.column = ['main_genres','count']
genre_countdf.sort_values(by='count',ascending=False,inplace=True)
genre_counttop = genre_countdf.head(10)
genre_counttop

fig,ax = plt.subplots(figsize=(8,10))
ax.pie(genre_counttop['count'],labels=genre_counttop['main_genres'],autopct='%1.1f%%')
plt.title('Number of Main Genres Produced')
plt.show()
```

```python
genre_profit = movies.groupby('main_genres')['Profit'].mean()
genre_profitdf = genre_profit.reset_index()
genre_profitdf.sort_values(by='Profit',ascending=False,inplace=True)
genre_profittop = genre_profitdf.head(10)
genre_profittop.plot(x='main_genres',y='Profit',figsize=(10,6),kind='bar')
plt.title('Average Profit by Main Genre')
plt.xlabel('Main Genre')
plt.ylabel('Average Profit')
plt.show()
```

```python
movies.head()
movies_correlation = movies[['budget','revenue','Profit']]
movies_correlation.corr()
sns.heatmap(movies_correlation,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
